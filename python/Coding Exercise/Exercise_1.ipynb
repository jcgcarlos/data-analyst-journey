## Sales Rep Performance Rating

You're analyzing sales performance for a BPO client. Write a program that takes a sales rep's monthly revenue and prints their performance rating.

Scenario
Given: monthly_revenue = 85000 Rules: >= 100000 → "Exceeds Target" >= 75000 → "Meets Target" >= 50000 → "Needs Improvement" below 50000 → "At Risk" Expected output: Meets Target

```python
monthly_revenue = 85000

# Write your if/elif/else here
if monthly_revenue >= 100000:
    print('Exceeds Target')
elif monthly_revenue >= 75000:
    print('Meets Target')
elif monthly_revenue >= 50000:
    print('Needs Improvement')
else:
    print('At Risk')
```

## SLA Breach Checker

You're monitoring call center SLA compliance. Check if a ticket's resolution time meets the SLA target and categorize it.

Scenario
Given: resolution_hours = 6 sla_target = 4 Rules: resolution_hours <= sla_target → "Within SLA" resolution_hours <= sla_target * 1.5 → "Minor Breach" anything above → "Major Breach" Expected output: Minor Breach

```python
resolution_hours = 6
sla_target = 4

# Write your if/elif/else here
if resolution_hours <= sla_target:
    print('Within SLA')
elif resolution_hours <= sla_target * 1.5: 
    print('Minor Breach')
else:
    print('Major Breach')
```

# Monthly Report Summary

You have a list of weekly ticket volumes for a month. Loop through the data and print whether each week was 'High Volume' (>500) or 'Normal Volume'.

Scenario
Given: weekly_tickets = [420, 610, 390, 540] Expected output: Week 1: Normal Volume Week 2: High Volume Week 3: Normal Volume Week 4: High Volume

```python
weekly_tickets = [420, 610, 390, 540]

for index, tickets in enumerate(weekly_tickets, start=1):
    if tickets > 500:
        print(f'Week {index}: High Volume')
    else:
        print(f'Week {index}: Normal Volume')
```

## Agent Score Classifier

You received QA scores for a team of agents. Loop through the scores and classify each one. Also count how many agents need coaching.

Scenario
Given: scores = [92, 68, 78, 55, 88, 71] Rules: >= 85 → "Excellent" >= 70 → "Good" below 70 → "Needs Coaching" Expected output: 92 → Excellent 68 → Needs Coaching 78 → Good 55 → Needs Coaching 88 → Excellent 71 → Good Agents needing coaching: 2

```python
scores = [92, 68, 78, 55, 88, 71]
coaching_count = 0

for i in scores:
    if i >= 85:
        print(f'{i} → Excellent')
    elif i >= 70:
        print(f'{i} → Good')
    else:
        print(f'{i} → Needs Coaching')
        coaching_count = coaching_count + 1 
print(f'Agents needing coaching: {coaching_count}')
    

```

## Revenue Trend Detector

You're building a simple trend detector. Given a list of monthly revenues, compare each month to the previous one and flag whether it went up, down, or stayed flat.

Scenario
Given: revenues = [50000, 62000, 58000, 62000, 75000] Expected output: Month 2: UP (+12000) Month 3: DOWN (-4000) Month 4: FLAT Month 5: UP (+13000)

```python
revenues = [50000, 62000, 58000, 62000, 75000]

for num in range(1, len(revenues)):
    diff = revenues[num] - revenues[num-1]
    if diff > 0:
        print(f"Month {num+1}: UP (+{diff})")
    elif diff < 0:
        print(f"Month {num+1}: DOWN ({diff})")
    else:
        print(f"Month {num+1}: FLAT")
```
